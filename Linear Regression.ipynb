This program will check the relation between heat sink temperatures of two modules in solar power plant using Linear Regression. The Data sets shows us the details on amount of solar irradiation and the Heat sink temperatures of the modules present in the power plant. 

```python
 %matplotlib inline
import pandas as pd
from pylab import *
```

The Data is loaded from the csv file through pandas

```python
data = pd.read_csv ("C:/Users/Bharath/Desktop/ML/L00144389/Data/Millenium_Energy.csv")
data
```

Now the available data from Heat sink temperature of Module 1 and Module 2 will be plotted using scatter function. 

```python
scatter(data.Mod1, data.Mod2)
```

Now the slope, intercept, r_value, p_value and standard error values will be computed using linear regression. With the r_value, r squared value is obtained

```python
from scipy import stats
```

```python
slope, intercept, r_value, p_value, std_error = stats.linregress(data.Mod1, data.Mod2)
r_value ** 2
```

Using Linear Regression we will define a function to draw a line using slope and intercept

```python
def line(value):
    return slope * value + intercept
```

```python
import matplotlib.pyplot as plt
```

Let's check if the line is fitting the data sets

```python
fig = plt.figure()
a = fig.add_subplot(1,1,1)
a.set_title('Linear Regression')

Line = line(data.Mod2)
plt.scatter(data.Mod1, data.Mod2)
plt.plot(data.Mod2, Line, c='r')
```

Since the Data points are fitting the line and r2 value is 0.987, Linear regression is suitable in this case.
